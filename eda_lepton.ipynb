```python
import numpy as np
import tensorflow as tf
import tensorflow.keras.backend as K
import matplotlib.pyplot as plt
```

```python
from process_data import make_ptregression_data

input_path = "data/train.npy"
arr = np.load(input_path)
```

```python
x_orig,x_smear,mll_smear = make_ptregression_data(arr)

orig_lep1_pt = x_orig[:,0]
orig_lep1_eta = x_orig[:,1]
orig_lep1_phi = x_orig[:,2]
orig_lep2_pt = x_orig[:,3]
orig_lep2_eta = x_orig[:,4]
orig_lep2_phi = x_orig[:,5]

smear_lep1_pt = x_smear[:,0]
smear_lep1_eta = x_smear[:,1]
smear_lep1_phi = x_smear[:,2]
smear_lep2_pt = x_smear[:,3]
smear_lep2_eta = x_smear[:,4]
smear_lep2_phi = x_smear[:,5]
```

```python
fig,ax = plt.subplots(3,2,figsize=(10,20))

ax[0,0].hist(orig_lep1_pt,bins=50,range=[0.,2.],histtype='step',density=1.,label='orig')
ax[0,0].hist(smear_lep1_pt,bins=50,range=[0.,2.],histtype='step',density=1.,label='smear')

ax[1,0].hist(orig_lep1_eta,bins=50,range=[-3.,3.],histtype='step',density=1.,label='orig')
ax[1,0].hist(smear_lep1_eta,bins=50,range=[-3.,3.],histtype='step',density=1.,label='smear')

ax[2,0].hist(orig_lep1_phi,bins=50,range=[-3.,3.],histtype='step',density=1.,label='orig')
ax[2,0].hist(smear_lep1_phi,bins=50,range=[-3.,3.],histtype='step',density=1.,label='smear')

ax[0,1].hist(orig_lep2_pt,bins=50,range=[0.,2.],histtype='step',density=1.,label='orig')
ax[0,1].hist(smear_lep2_pt,bins=50,range=[0.,2.],histtype='step',density=1.,label='smear')

ax[1,1].hist(orig_lep2_eta,bins=50,range=[-3.,3.],histtype='step',density=1.,label='orig')
ax[1,1].hist(smear_lep2_eta,bins=50,range=[-3.,3.],histtype='step',density=1.,label='smear')

ax[2,1].hist(orig_lep2_phi,bins=50,range=[-3.,3.],histtype='step',density=1.,label='orig')
ax[2,1].hist(smear_lep2_phi,bins=50,range=[-3.,3.],histtype='step',density=1.,label='smear')
```
